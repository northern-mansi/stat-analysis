```python
import os
import pandas as pd
```

```python
files = os.listdir("./src")
files = [os.path.join("./src", fn) for fn in files if fn.endswith(".csv") and fn.startswith("NM")]
files
```

```python
# Quick check for consistency
types = set()
for fn in files:
    cols = set(list(pd.read_csv(fn).columns))
    cols = cols - {"n", "Text", "Place", "Comment", "Unnamed: 4", "Unnamed: 3", "Unnamed: 0", "No"}
    types.add(tuple(cols))
    if "SenT" not in cols:
        print(f"{fn} does not contain SenT")
for typ in types:
    print(typ)
    print()
```

```python
merged_df = None
for e in range(len(files)):
    if e == 3:
        continue
    cur_df = pd.read_csv(files[e])
    cur_df = cur_df.drop(columns=["n", "Text", "Place", "Comment", "No"], errors="ignore")
    cur_df = cur_df.rename(columns={"Unnamed: 0": "Sentence"})
    cur_df["File"] = files[e]
    if e == 0:
        merged_df = cur_df
    else:
        merged_df = pd.concat([merged_df, cur_df])
merged_df
```

```python
df4 = pd.read_csv(files[3])
df4 = df4.drop(columns=["n", "Text", "Place", "Comment", "No", "Unnamed: 4"], errors="ignore")
df4 = df4.drop(index=0)
df4 = df4.rename(columns={"Unnamed: 3": "Sentence"})
df4["File"] = files[3]
df4
```
